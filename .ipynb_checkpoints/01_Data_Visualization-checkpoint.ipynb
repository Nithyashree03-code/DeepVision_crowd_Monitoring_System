```python
import os
import cv2
import random
import matplotlib.pyplot as plt

dataset_path = r"C:\Users\nshre\OneDrive\Desktop\DeepVision\dataset\crowd"

valid_ext = [".jpg", ".jpeg", ".png"]
images = [f for f in os.listdir(dataset_path) if f.lower().endswith(tuple(valid_ext))]

print("Total valid images found:", len(images))

if len(images) == 0:
    print(" ERROR: No image files found!")
else:
    first_img_path = os.path.join(dataset_path, images[0])
    print("First image path:", first_img_path)
    img = cv2.imread(first_img_path)
    if img is None:
        print(" ERROR: OpenCV could not load this image!")
    else:
        # Convert BGR  RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Display the first image
        plt.figure(figsize=(5,5))
        plt.imshow(img)
        plt.title("Sample Image")
        plt.axis("off")
        plt.show()
        plt.figure(figsize=(10,10))

        # Random selection (max 9)
        random_imgs = random.sample(images, min(9, len(images)))

        for i, img_name in enumerate(random_imgs):
            img_path = os.path.join(dataset_path, img_name)
            img = cv2.imread(img_path)

            if img is None:
                continue

            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            plt.subplot(3, 3, i+1)
            plt.imshow(img)
            plt.title(img_name[:15])
            plt.axis("off")

        plt.suptitle("Random Dataset Visualization", fontsize=16)
        plt.show()
```
