```python
import os
import cv2
import numpy as np
from tqdm import tqdm
import random

# Paths
dataset_path = r"C:\Users\nshre\OneDrive\Desktop\DeepVision\dataset\crowd"
output_path = r"C:\Users\nshre\OneDrive\Desktop\DeepVision\dataset\crowd_preprocessed"
os.makedirs(output_path, exist_ok=True)

# Allowed image formats
valid_ext = [".jpg", ".jpeg", ".png"]

# List image files
images = [f for f in os.listdir(dataset_path) if f.lower().endswith(tuple(valid_ext))]
print("Total images found:", len(images))

# Resize dimensions
IMG_HEIGHT = 224
IMG_WIDTH = 224

# Helper functions for augmentation
def random_flip(img):
    if random.random() < 0.5:
        img = cv2.flip(img, 1)
    return img

def random_rotate(img, max_angle=20):
    if random.random() < 0.5:
        angle = random.uniform(-max_angle, max_angle)
        M = cv2.getRotationMatrix2D((IMG_WIDTH/2, IMG_HEIGHT/2), angle, 1)
        img = cv2.warpAffine(img, M, (IMG_WIDTH, IMG_HEIGHT), borderMode=cv2.BORDER_REFLECT)
    return img

def random_brightness_contrast(img, brightness=0.2, contrast=0.2):
    if random.random() < 0.5:
        b = random.uniform(-brightness, brightness)
        c = random.uniform(1-contrast, 1+contrast)
        img = np.clip(c * img + b*255, 0, 255).astype(np.uint8)
    return img

# List to store all preprocessed images for .npy
X = []

count = 0
for img_name in tqdm(images, desc="Preprocessing Images"):
    img_path = os.path.join(dataset_path, img_name)
    
    # Load image
    img = cv2.imread(img_path)
    if img is None:
        print(f"Skipping corrupted: {img_name}")
        continue

    # Resize
    img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

    # Convert BGR → RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Apply augmentations
    img = random_flip(img)
    img = random_rotate(img)
    img = random_brightness_contrast(img)

    # Normalize and standardize
    img_norm = img / 255.0
    img_norm = (img_norm - np.mean(img_norm)) / (np.std(img_norm) + 1e-7)

    # Append to list for .npy
    X.append(img_norm)

    # Convert back to [0,255] uint8 for saving as image
    save_img = ((img_norm - img_norm.min()) / (img_norm.max() - img_norm.min()) * 255).astype(np.uint8)
    save_path = os.path.join(output_path, f"preprocessed_{count}.png")
    cv2.imwrite(save_path, cv2.cvtColor(save_img, cv2.COLOR_RGB2BGR))

    count += 1

# Convert list to numpy array and save as .npy
X = np.array(X, dtype=np.float32)
np.save("X_images_preprocessed.npy", X)

print(f"Preprocessing Completed! {count} images saved to {output_path}")
print(f"All preprocessed images also saved in X_images_preprocessed.npy with shape {X.shape}")
```

```python
import os
import cv2
import matplotlib.pyplot as plt
import random

# Path to preprocessed images folder
folder = r"C:\Users\nshre\OneDrive\Desktop\DeepVision\dataset\crowd_preprocessed"

# List all images
images = [f for f in os.listdir(folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Pick random images
num_images = 9
sample_images = random.sample(images, min(num_images, len(images)))

plt.figure(figsize=(10,10))

for i, img_name in enumerate(sample_images):
    img_path = os.path.join(folder, img_name)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    
    plt.subplot(3,3,i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(img_name)

plt.tight_layout()
plt.show()
```
